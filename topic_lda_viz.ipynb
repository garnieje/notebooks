```python
import pyLDAvis.gensim
import gensim
```

```python
PATH_DICTIONARY = "/Users/jerome/Documents/influencer/topics/tmp_dictionary.bz2"
PATH_CORPUS = "/Users/jerome/Documents/influencer/topics/tmp_corpus.mm"
PATH_MODEL = "/Users/jerome/Documents/influencer/topics/tmp_lda"

dictionary = gensim.corpora.Dictionary.load_from_text(PATH_DICTIONARY)
corpus = gensim.corpora.MmCorpus(PATH_CORPUS)
lda = gensim.models.ldamodel.LdaModel.load(PATH_MODEL)
```

```python
display_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(display_data)
```

```python
tsne_display_data = pyLDAvis.prepare(lda, corpus, dictionary, mds='tsne')
pyLDAvis.display(tsne_display_data)
```
